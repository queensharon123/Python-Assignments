```python
np.random.seed(42) 
df['height'] = np.random.randint(150, 181, size=len(df))
print(df.head()) 
```

```python
if 'Team' in df.columns:
    team_column = 'Team'

team_distribution = df[team_column].value_counts()


total_employees = len(df)

percentage_split = (team_distribution / total_employees) * 100


team_distribution_df = pd.DataFrame({
    'Employee Count': team_distribution,
    'Percentage Split (%)': percentage_split
})


print(team_distribution_df)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

team_distribution = df[team_column].value_counts()


team_distribution_df = pd.DataFrame({
    'Employee Count': team_distribution,
    'Percentage Split (%)': (team_distribution / len(df)) * 100
})


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1) 
sns.barplot(x=team_distribution_df.index, y='Employee Count', data=team_distribution_df, palette="viridis")
plt.title('Employee Distribution by Team')
plt.xlabel('Team')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
```

```python
# 2

if 'Position' in df.columns:
    position_column = 'Position'

position_counts = df[position_column].value_counts()

total_employees = len(df)

position_percentage = (position_counts / total_employees) * 100


position_distribution_df = pd.DataFrame({
    'Employee Count': position_counts,
    'Percentage (%)': position_percentage
})


print(position_distribution_df)
```

```python

position_counts = df[position_column].value_counts()


total_employees = len(df)


position_percentage = (position_counts / total_employees) * 100

position_distribution_df = pd.DataFrame({
    'Employee Count': position_counts,
    'Percentage (%)': position_percentage
})


plt.figure(figsize=(14, 6))


plt.subplot(1, 2, 1)  
sns.barplot(x=position_distribution_df.index, y='Employee Count', data=position_distribution_df, palette="viridis")
plt.title('Employee Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Employee Count')
plt.xticks(rotation=45, ha='right')  
```

```python
# 3

bins = [20, 30, 40, 50, 60, 70] 
labels = ['20-29', '30-39', '40-49', '50-59', '60-69']  
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)  

age_group_counts = df['Age Group'].value_counts().sort_index()

predominant_age_group = age_group_counts.idxmax()
predominant_count = age_group_counts.max()
```

```python

plt.figure(figsize=(10, 5))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette="viridis")
plt.title('Employee Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Employee Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

highest_expenditure_row = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]

highest_team = highest_expenditure_row['Team']
highest_position = highest_expenditure_row['Position']
highest_salary_expenditure = highest_expenditure_row['Salary']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=salary_expenditure, x='Salary', y='Position', hue='Team', dodge=True, palette='viridis')
plt.title('Salary Expenditure by Team and Position')
plt.xlabel('Total Salary Expenditure ($)')
plt.ylabel('Position')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 5

print(df[['Age', 'Salary']].isnull().sum())

df_clean = df.dropna(subset=['Age', 'Salary'])

correlation = df_clean['Age'].corr(df_clean['Salary'])
print(f"The correlation coefficient between Age and Salary is: {correlation:.2f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='Age', y='Salary', alpha=0.6)

sns.regplot(data=df_clean, x='Age', y='Salary', scatter=False, color='red')

plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.show()
```
